```
# csv to DataFrame

# 사전에 랭킹 순으로 
# 해시태그 내에 한글이 포함된 것들을
# 순위, 해시태그, 사용 수
# 순서대로 저장해 둠

import csv
import pandas as pd

# tagRanking.csv 파일
filePath = "/content/drive/My Drive/Colab Notebooks/T3 Insta Profiling/tagRanking.csv"
DF = pd.read_csv(filePath)

DF
```

```
# 해시태그를 
# 한 글자씩 쪼개서 저장
# 생각해보니 set 쓰면 되서 필요없음ㅋ

word = []
for idx, hashtag in enumerate(DF["tag"]):
    word.append([w for w in hashtag])

DF["t_tag"] = word
```

## 유사도 계산

간단한 유사도 계산을 함

(겹치는 글자 수) / ((해시태그1의 길이) + (해시태그2의 길이) + 1) 

분모의 1은 DivideByZero 방지

## 해시태그 전처리 01

해시태그 내에 필요 없다고 판단되는 다음 것들을 제거 후,

(#,인스타그램,스타그램,인스타,스타,그램)






```
# 해시태그 전처리 01

def preprocessingTag(text):
    # #기호는 모든 해시태그에 포함되어 유사도 계산에 크게 영향을 줌
    text = text.replace("#","")
    # 쓰잘데기 없는 것들
    text = text.replace("인스타그램","")
    text = text.replace("스타그램","")
    text = text.replace("인스타","")
    text = text.replace("스타","")
    text = text.replace("그램","")
    # 기타 고려해볼 대상
    # 맞팔, 인친, 

    return text

DF["p_tag"] = DF["tag"].apply(preprocessingTag)

# 해시태그 클러스터링
DF["cluster"] = ""

for idx1, hashtag1 in enumerate(DF["p_tag"]):
    tmp = []
    for idx2, hashtag2 in enumerate(DF["p_tag"]):
        
        # 같은건 패스
        if idx1 == idx2:
            continue

        # 교집합
        intersec = list(set(hashtag1) & set(hashtag2))
        # 유사도 점수
        similarity = len(intersec) / (len(hashtag1)+len(hashtag2)+1)
        # 특정 점수 구간 추출
        if 0.33 < similarity < 0.75:
            print(DF["tag"][idx1], DF["tag"][idx2])
            tmp.append(DF["tag"][idx2])

    DF["cluster"][idx1] = tmp
```

## 해시태그 전처리 02

해시태그 내에 필요 없다고 판단되는 다음 것들을 추가,

지명으로 세분화 되는 해시태그 (ex 안양맛집, 강남카페, 수원맘, 도쿄여행 등)

이런 지명을 '맛집'이라는 단어가 포함되는 해시태그들을 통해서 추출하고자 함

('맛집' 외에도 '카페','여행' 등의 단어도 고려 가능)



```
# 해시태그 전처리 02

# 해시태그 끼리 중복되는 내용 & 
# 비슷한 의미를 지닌 것들이 너무 많아 제대로 쓰기 어려움

# 그래서 지역 만큼은 전처리를 하고자 함
# '맛집'이라는 키워드가 들어간 경우, 지명이 함께 나올 확률이 높다고 생각해서
# 이를 이용해 해시태그에 사용된 지명들을 추출하고자 함
# 근데 지명이랑 음식도 딸려나옴
# 일단 진행

# 또 다른 후보로는
# '카페', '여행'

region = []

for hashtag in DF["tag"]:
    if "맛집" in hashtag:
        region.append(hashtag.replace("맛집","").replace("#",""))
```

```
print(region)
```

```
# 해시태그 전처리 02

def preprocessingTag(text):
    # #기호는 모든 해시태그에 포함되어 유사도 계산에 크게 영향을 줌
    text = text.replace("#","")
    # 쓰잘데기 없는 것들
    text = text.replace("인스타그램","")
    text = text.replace("스타그램","")
    text = text.replace("인스타","")
    text = text.replace("스타","")
    text = text.replace("그램","")
    # 지명,음식명 제거
    for word in region:
        text = text.replace(word,"")

    return text

DF["pg_tag"] = DF["tag"].apply(preprocessingTag)

# 해시태그 클러스터링
DF["cluster_NoRegion"] = ""

for idx1, hashtag1 in enumerate(DF["pg_tag"]):
    tmp = []
    for idx2, hashtag2 in enumerate(DF["pg_tag"]):
        
        # 같은건 패스
        if idx1 == idx2:
            continue

        # 교집합
        intersec = list(set(hashtag1) & set(hashtag2))
        # 유사도 점수
        similarity = len(intersec) / (len(hashtag1)+len(hashtag2)+1)
        # 특정 점수 구간 추출
        if 0.33 < similarity < 0.75:
            print(DF["tag"][idx1], DF["tag"][idx2])
            tmp.append(DF["tag"][idx2])

    DF["cluster_NoRegion"][idx1] = tmp
```

```
# tagCluster.csv 파일 저장
# rank     tag      toal    t_tag  p_tag    cluster         pg_tag   cluster_NoRegion
# 사용랭킹 해시태그 사용수  글자별 전처리01 p_tag기반군집화 전처리02 pg_tag기반군집화

filePath = "/content/drive/My Drive/Colab Notebooks/T3 Insta Profiling/tagCluster.csv"
DF.to_csv(filePath, encoding='utf-8-sig')
```

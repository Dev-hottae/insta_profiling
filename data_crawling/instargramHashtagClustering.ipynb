## í•´ì‹œíƒœê·¸ ì „ì²˜ë¦¬ 01

í•´ì‹œíƒœê·¸ ë‚´ì— í•„ìš” ì—†ë‹¤ê³  íŒë‹¨ë˜ëŠ” ë‹¤ìŒ ê²ƒë“¤ì„ ì œê±° í›„,

(#,ì¸ìŠ¤íƒ€ê·¸ë¨,ìŠ¤íƒ€ê·¸ë¨,ì¸ìŠ¤íƒ€,ìŠ¤íƒ€,ê·¸ë¨)






```
# í•´ì‹œíƒœê·¸ ì „ì²˜ë¦¬ 01

def preprocessingTag(text):
    # #ê¸°í˜¸ëŠ” ëª¨ë“  í•´ì‹œíƒœê·¸ì— í¬í•¨ë˜ì–´ ìœ ì‚¬ë„ ê³„ì‚°ì— í¬ê²Œ ì˜í–¥ì„ ì¤Œ
    text = text.replace("#","")
    # ì“°ì˜ë°ê¸° ì—†ëŠ” ê²ƒë“¤
    text = text.replace("ì¸ìŠ¤íƒ€ê·¸ë¨","")
    text = text.replace("ìŠ¤íƒ€ê·¸ë¨","")
    text = text.replace("ì¸ìŠ¤íƒ€","")
    text = text.replace("ìŠ¤íƒ€","")
    text = text.replace("ê·¸ë¨","")
    # ê¸°íƒ€ ê³ ë ¤í•´ë³¼ ëŒ€ìƒ
    # ë§íŒ”, ì¸ì¹œ, 

    return text

DF["p_tag"] = DF["tag"].apply(preprocessingTag)

# í•´ì‹œíƒœê·¸ í´ëŸ¬ìŠ¤í„°ë§
DF["cluster"] = ""

for idx1, hashtag1 in enumerate(DF["p_tag"]):
    tmp = []
    for idx2, hashtag2 in enumerate(DF["p_tag"]):
        
        # ê°™ì€ê±´ íŒ¨ìŠ¤
        if idx1 == idx2:
            continue

        # êµì§‘í•©
        intersec = list(set(hashtag1) & set(hashtag2))
        # ìœ ì‚¬ë„ ì ìˆ˜
        similarity = len(intersec) / (len(hashtag1)+len(hashtag2)+1)
        # íŠ¹ì • ì ìˆ˜ êµ¬ê°„ ì¶”ì¶œ
        if 0.33 < similarity < 0.75:
            print(DF["tag"][idx1], DF["tag"][idx2])
            tmp.append(DF["tag"][idx2])

    DF["cluster"][idx1] = tmp
```

ì´ëŸ° ì§€ëª…ì„ 'ë§›ì§‘'ì´ë¼ëŠ” ë‹¨ì–´ê°€ í¬í•¨ë˜ëŠ” í•´ì‹œíƒœê·¸ë“¤ì„ í†µí•´ì„œ ì¶”ì¶œí•˜ê³ ì í•¨

('ë§›ì§‘' ì™¸ì—ë„ 'ì¹´í˜','ì—¬í–‰' ë“±ì˜ ë‹¨ì–´ë„ ê³ ë ¤ ê°€ëŠ¥)



```
# í•´ì‹œíƒœê·¸ ì „ì²˜ë¦¬ 02

# í•´ì‹œíƒœê·¸ ë¼ë¦¬ ì¤‘ë³µë˜ëŠ” ë‚´ìš© & 
# ë¹„ìŠ·í•œ ì˜ë¯¸ë¥¼ ì§€ë‹Œ ê²ƒë“¤ì´ ë„ˆë¬´ ë§ì•„ ì œëŒ€ë¡œ ì“°ê¸° ì–´ë ¤ì›€

# ê·¸ë˜ì„œ ì§€ì—­ ë§Œí¼ì€ ì „ì²˜ë¦¬ë¥¼ í•˜ê³ ì í•¨
# 'ë§›ì§‘'ì´ë¼ëŠ” í‚¤ì›Œë“œê°€ ë“¤ì–´ê°„ ê²½ìš°, ì§€ëª…ì´ í•¨ê»˜ ë‚˜ì˜¬ í™•ë¥ ì´ ë†’ë‹¤ê³  ìƒê°í•´ì„œ
# ì´ë¥¼ ì´ìš©í•´ í•´ì‹œíƒœê·¸ì— ì‚¬ìš©ëœ ì§€ëª…ë“¤ì„ ì¶”ì¶œí•˜ê³ ì í•¨
# ê·¼ë° ì§€ëª…ì´ë‘ ìŒì‹ë„ ë”¸ë ¤ë‚˜ì˜´
# ì¼ë‹¨ ì§„í–‰

# ë˜ ë‹¤ë¥¸ í›„ë³´ë¡œëŠ”
# 'ì¹´í˜', 'ì—¬í–‰'

region = []

for hashtag in DF["tag"]:
    if "ë§›ì§‘" in hashtag:
        region.append(hashtag.replace("ë§›ì§‘","").replace("#",""))
```

```
print(region)
```

```
# í•´ì‹œíƒœê·¸ ì „ì²˜ë¦¬ 02

def preprocessingTag(text):
    # #ê¸°í˜¸ëŠ” ëª¨ë“  í•´ì‹œíƒœê·¸ì— í¬í•¨ë˜ì–´ ìœ ì‚¬ë„ ê³„ì‚°ì— í¬ê²Œ ì˜í–¥ì„ ì¤Œ
    text = text.replace("#","")
    # ì“°ì˜ë°ê¸° ì—†ëŠ” ê²ƒë“¤
    text = text.replace("ì¸ìŠ¤íƒ€ê·¸ë¨","")
    text = text.replace("ìŠ¤íƒ€ê·¸ë¨","")
    text = text.replace("ì¸ìŠ¤íƒ€","")
    text = text.replace("ìŠ¤íƒ€","")
    text = text.replace("ê·¸ë¨","")
    # ì§€ëª…,ìŒì‹ëª… ì œê±°
    for word in region:
        text = text.replace(word,"")

    return text

DF["pg_tag"] = DF["tag"].apply(preprocessingTag)

# í•´ì‹œíƒœê·¸ í´ëŸ¬ìŠ¤í„°ë§
DF["cluster_NoRegion"] = ""

for idx1, hashtag1 in enumerate(DF["pg_tag"]):
    tmp = []
    for idx2, hashtag2 in enumerate(DF["pg_tag"]):
        
        # ê°™ì€ê±´ íŒ¨ìŠ¤
        if idx1 == idx2:
            continue

        # êµì§‘í•©
        intersec = list(set(hashtag1) & set(hashtag2))
        # ìœ ì‚¬ë„ ì ìˆ˜
        similarity = len(intersec) / (len(hashtag1)+len(hashtag2)+1)
        # íŠ¹ì • ì ìˆ˜ êµ¬ê°„ ì¶”ì¶œ
        if 0.33 < similarity < 0.75:
            print(DF["tag"][idx1], DF["tag"][idx2])
            tmp.append(DF["tag"][idx2])

    DF["cluster_NoRegion"][idx1] = tmp
```

```
# Crawling
from bs4 import BeautifulSoup
import requests
import re

i=0
while(True):

    URL = "https://top-hashtags.com/instagram/{}".format(i*100+1)

    response = requests.get(URL)
    soup = BeautifulSoup(response.text, "html.parser") 

    rank=[]
    tmp = soup.select('ul.i-group div.i-num')
    rank = re.findall("[0-9]+",str(tmp))
    
    tag=[]
    tmp = soup.select('ul.i-group div.i-tag')
    tag = re.findall("#[^ ]+(?=</a>)",str(tmp))

    total=[]
    tmp = soup.select('ul.i-group div.i-total')
    total = re.findall("(?<=total\">)[^ ]*(?=</div>)",str(tmp))
    

    for idx in range(100):
        isKor = re.findall("[가-힣].",tag[idx])
        if len(isKor) != 0:
            print(rank[idx],tag[idx],total[idx])
    i+=1
```

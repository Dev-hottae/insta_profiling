```
# Crawling
from bs4 import BeautifulSoup
import requests
import re
# Data Store
import json
import csv

i=0
# temp json data 
data = {}
data['ranking'] = []

# temp csv data
csv_file_path = "/content/drive/My Drive/insta/tagRanking.csv"
with open(csv_file_path,'a',newline='') as data_csv:
    write_data_CSV = csv.writer(data_csv) 
    tmp_list = []
    tmp_list.append("rank")
    tmp_list.append("tag")
    tmp_list.append("total")
    write_data_CSV.writerow(tmp_list)


while(True):

    URL = "https://top-hashtags.com/instagram/{}".format(i*100+1)

    response = requests.get(URL)
    soup = BeautifulSoup(response.text, "html.parser") 

    rank=[]
    tmp = soup.select('ul.i-group div.i-num')
    rank = re.findall("[0-9]+",str(tmp))
    
    tag=[]
    tmp = soup.select('ul.i-group div.i-tag')
    tag = re.findall("#[^ ]+(?=</a>)",str(tmp))

    total=[]
    tmp = soup.select('ul.i-group div.i-total')
    total = re.findall("(?<=total\">)[^ ]*(?=</div>)",str(tmp))

    for idx in range(100):
        isKor = re.findall("[가-힣].",tag[idx])
        if len(isKor) != 0:
            data['ranking'].append({
                'rank' : rank[idx],
                'tag' : tag[idx],
                'total' : total[idx]
            })
            # save as csv
            with open(csv_file_path,'a',newline='') as data_csv:
                write_data_CSV = csv.writer(data_csv) 
                tmp_list = []
                tmp_list.append(rank[idx])
                tmp_list.append(tag[idx])
                tmp_list.append(total[idx])
                write_data_CSV.writerow(tmp_list)
            
            print(rank[idx],tag[idx],total[idx])

    if i%100 == 99:
        
        json_file_path = "/content/drive/My Drive/insta/tagRanking{}.json".format(i//100)
        # save as json
        with open(json_file_path, 'w') as outfile:
            json.dump(data, outfile)
            data = {}
            data['ranking'] = []
    i+=1
```
